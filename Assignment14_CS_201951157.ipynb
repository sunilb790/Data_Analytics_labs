```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/sample_data/IPL_Auction_2022_FullList.csv')
data.head()
```

```python
table=pd.crosstab(data["Specialism"],data["Batting"],margins=True)
table
```

```python
table=pd.crosstab(data["Specialism"],data["Batting"],margins=False)
table
```

```python
import seaborn as sns
from scipy import stats

df=3
chi_data = stats.chi2.rvs(table)
sns.histplot(chi_data,kde=True)
```

```python
from scipy import stats

chisq, pvalue, df, expected = stats.chi2_contingency(table)
print(f'Observed chi2: {chisq:.4f}')
print(f'p-value: {pvalue:.4f}')
```
